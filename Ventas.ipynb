```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Alejandro Navas\Downloads\Online Sales Data.csv")
df.head(10)
```

```python
## categoria m√°s comprada por region ##
```

```python
plt.style.use('fivethirtyeight')
color = sns.color_palette('husl')
region_sales = df.groupby('Region')['Total Revenue'].sum().sort_values(ascending=True)
region_sales.plot(kind='bar', figsize=(10, 6),color = color)
plt.title('Ventas totales por regi√≥n')
plt.xlabel('Regi√≥n')
plt.ylabel('Ventas totales')
plt.xticks(rotation=45)
plt.show()
```

```python
##An√°lisis de ventas por categor√≠a de producto
category_sales = df.groupby('Product Category')['Total Revenue'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', figsize=(10, 6),color=color)
plt.title('Ganancias totales por categor√≠a de producto')
plt.xlabel('Categor√≠a de producto')
plt.ylabel('Ventas totales')
plt.show()
```

```python
##An√°lisis de productos m√°s vendidos
top_products = df.groupby('Product Name')['Units Sold'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', figsize=(10, 6),color=color)
plt.title('Productos m√°s vendidos')
plt.xlabel('Unidades vendidas')
plt.ylabel('Nombre del producto')
plt.show()
```

```python
##An√°lisis de tendencias de ventas en el tiempo
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(pd.Grouper(key='Date', freq='M'))['Total Revenue'].sum()
monthly_sales.plot(figsize=(12, 6),color=color)
plt.title('Ventas mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ventas totales')
plt.show()
```

```python
## An√°lisis de m√©todos de pago
payment_methods = df['Payment Method'].value_counts()
payment_methods.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('M√©todos de pago')
plt.axis('equal')
plt.show()
```
